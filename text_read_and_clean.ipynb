```python
import pandas as pd
from collections import Counter
import re

def count_words(texts):
    word_counts = Counter()
    all_words = []

    # Count the occurrences of each word in each text
    for text in texts:
        # Use regex to split text into words (ignores punctuation and case)
        words = re.findall(r'\b\w+\b', text.lower())
        word_counts.update(words)
        all_words.extend(words)

    # Remove duplicates while maintaining order
    all_words = list(dict.fromkeys(all_words))

    # Create DataFrame with word counts
    df = pd.DataFrame(columns=all_words)

    # Add a row for each text
    for i, text in enumerate(texts):
        word_counts_text = Counter(re.findall(r'\b\w+\b', text.lower()))
        row = {word: word_counts_text[word] for word in all_words}
        df.loc[i] = row

    return df

texts = ["text1", "text2", "text3"]  # Replace with the list of texts
word_count_df = count_words(textList)
word_count_df








```
